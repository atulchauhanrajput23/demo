## .apply() Method with Row Values in Pandas

```python
import pandas as pd

bond = pd.read_csv(r"C:\Users\umang\Data Science\Data Sets\jamesbond.csv",index_col='Film')

bond.sort_index(inplace=True)

bond.head()
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    
    else:
        return "I have no Idea"
```

```python
bond.apply(good_movie,axis="columns")
```

```python
bond['review'] = bond.apply(good_movie,axis="columns")
```
